## Ordinal Number Encoding

```python
import datetime
```

```python
todaydate=datetime.datetime.today()
```

```python
print(todaydate)
```

```python
todaydate-datetime.timedelta(3) # This will give 3 days back date
```

```python
# Using List Comprehension to calculate days:

List=[todaydate-datetime.timedelta(x) for x in range(0,15)]
```

```python
List
```

```python
import pandas as pd
data=pd.DataFrame(List)
data.columns=["Day"]
```

```python
data.head()
```

```python
data['weekday']=data['Day'].dt.day_name()
data.head()
```

```python
dictionary={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
```

```python
data['weekday_ordinal']=data['weekday'].map(dictionary)
```

#### Count Or Frequency Encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
```

```python
train_set=train_set[columns]
```

```python
train_set
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set
```

```python
train_set['Employment'].value_counts()
```

```python
for feature in train_set.columns[:]:
        print(feature,":",len(train_set[feature].unique()),'labels') # To find the labels count
```

```python
country_map=train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

# Target Guided ordinal Encoding 

```python
import pandas as pd
df=pd.read_csv('titanic.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()# probability of person surviving
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)} # We are assigning the values on the value of ranks
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

#### Mean Encoding

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```
